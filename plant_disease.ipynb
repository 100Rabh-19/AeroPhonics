```python
import os
from PIL import Image
import imagehash

database_path = "winnig_dataset"
hash_db = []

for label in os.listdir(database_path):
    class_path = os.path.join(database_path, label)
    
    for img_name in os.listdir(class_path):
        img_path = os.path.join(class_path, img_name)
        
        try:
            img = Image.open(img_path)
            h = imagehash.phash(img)
            
            hash_db.append((h, label))
        except:
            print("Error:", img_path)

print("Total images processed:", len(hash_db))
```

```python
def predict_image(img_path):
    from PIL import Image
    import imagehash
    
    img = Image.open(img_path)
    input_hash = imagehash.phash(img)
    
    best_match = None
    min_diff = float("inf")
    
    for h, label in hash_db:
        diff = input_hash - h
        
        if diff < min_diff:
            min_diff = diff
            best_match = label
    
    # threshold
    if min_diff <= 5:
        confidence = max(0, 100 - min_diff * 10)
        return best_match, confidence
    else:
        return "Unknown", 0
```

```python
label, conf = predict_image("C:/100rabh/winnig_dataset/CedarRust/AppleCedarRust3.JPG")
print(f"Prediction: {label}, Confidence: {conf}%")
```
